```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

---
Fuzziyfikasi Banyak Pembeli
---

banyak pembeli memiliki rentang 1-30 orang:

```python
banyak_pembeli = ctrl.Antecedent(np.arange(0,30,1),'banyak_pembeli')
pembeli_x = np.arange(0,30,1) #untuk plotting
```

```python
banyak_pembeli['low'] = fuzz.trapmf(banyak_pembeli.universe,[0,0,1,10])
banyak_pembeli['medium'] = fuzz.trimf(banyak_pembeli.universe,[8,11,14])
banyak_pembeli['high'] = fuzz.trapmf(banyak_pembeli.universe,[13,20,30,30])
```

---
plotting membership function "banyak pembeli"
---

plotting fungsi keanggotaan low

```python
pembeli_low = fuzz.trapmf(pembeli_x, [0,0,1,10])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(pembeli_x,pembeli_low,'blue')
ax0.set_title('pembeli low')
```

plotting fungsi keanggotaan medium

```python
pembeli_medium = fuzz.trimf(pembeli_x, [8,11,14])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(pembeli_x,pembeli_medium,'red')
ax0.set_title('pembeli medium')
```

plotting fungsi keanggotaan high

```python
pembeli_high = fuzz.trapmf(pembeli_x, [13,20,30,30])
fig, (ax0) = plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(pembeli_x,pembeli_high,'green')
ax0.set_title('pembeli high')
```

plotting ke tiga fungsi keanggotaan banyak pembeli

```python
banyak_pembeli.view()
```

Fuzzifikasi Tingkat Kelarisan
---

```python
kelarisan = ctrl.Consequent(np.arange(0,100,1),'laris')
kelarisan_x = np.arange(0,100,1)
```

```python
kelarisan['low'] = fuzz.trapmf(kelarisan.universe,[0,0,1,30])
kelarisan['medium'] = fuzz.trimf(kelarisan.universe,[30,50,70])
kelarisan['high'] = fuzz.trimf(kelarisan.universe,[70,100,100])
```

---
Plotting Membership Function "Tingkat Kelarisan"
---

```python
kelarisan_low = fuzz.trapmf(kelarisan_x,[0,0,1,30])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(kelarisan_x,kelarisan_low,'blue')
ax0.set_title('low')
```

```python
kelarisan_medium = fuzz.trimf(kelarisan_x,[30,50,70])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(kelarisan_x,kelarisan_medium,'orange')
ax0.set_title('medium')
```

```python
kelarisan_high = fuzz.trimf(kelarisan_x,[70,100,100])
fig,(ax0)= plt.subplots(nrows=1,figsize=(6,6))
ax0.plot(kelarisan_x,kelarisan_high,'green')
ax0.set_title('high')
```

```python
kelarisan.view()
```

---
inferensi
---

```python
rule1 = ctrl.Rule(banyak_pembeli['low'] & uang_masuk['low'], kelarisan['low'])
rule2 = ctrl.Rule(banyak_pembeli['low'] & uang_masuk['medium'], kelarisan['low'])
rule3 = ctrl.Rule(banyak_pembeli['low'] & uang_masuk['high'], kelarisan['medium'])

rule4 = ctrl.Rule(banyak_pembeli['medium'] & uang_masuk['low'], kelarisan['low'])
rule5 = ctrl.Rule(banyak_pembeli['medium'] & uang_masuk['medium'], kelarisan['medium'])
rule6 = ctrl.Rule(banyak_pembeli['medium'] & uang_masuk['high'], kelarisan['high'])

rule7 = ctrl.Rule(banyak_pembeli['high'] & uang_masuk['low'], kelarisan['medium'])
rule8 = ctrl.Rule(banyak_pembeli['high'] & uang_masuk['medium'], kelarisan['high'])
rule9 = ctrl.Rule(banyak_pembeli['high'] & uang_masuk['high'], kelarisan['high'])
```

---
Rule Evaluation
---

```python
laris_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
```

---
Defuzzifikasi
---

```python
laris = ctrl.ControlSystemSimulation(laris_ctrl)
```

```python
laris.input['banyak_pembeli'] = 9
laris.input['uang_masuk'] = 286
```

```python
laris.compute()
```

```python
print(laris.output['laris'])
```

```python
banyak_pembeli.view(sim=laris)
```

```python
uang_masuk.view(sim=laris)
```

```python
kelarisan.view(sim=laris)
```
